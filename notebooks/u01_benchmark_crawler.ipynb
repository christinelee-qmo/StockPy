### 下載某天的資料

```python
import requests

import datetime

date = datetime.date(2019,1,2)
datestr = date.strftime("%Y%m%d") #改年月日的格式(沒有dash)

# http://www.twse.com.tw/exchangeReport/MI_5MINS_INDEX?response=csv&date=20190102&_=1552616075051
res = requests.get("http://www.twse.com.tw/exchangeReport/MI_5MINS_INDEX?response=csv&date="+datestr+"&_=1552616075051")

res.text[:1000]
```

### 將資料用pandas整理成表格(dataframe)

Pandas 教學：
http://pandas.pydata.org/pandas-docs/stable/getting_started/10min.html

```python
import pandas as pd
import io #虛擬？檔案
#df = pd.read_csv(io.StringIO(res.text)) 匯入成CSV
#.text.replace("","")
df = pd.read_csv(io.StringIO(res.text.replace("=", "")), header=1, index_col='時間') #header 設定第一欄為標籤，時間列為第一列

df[:10]
```

## 資料處理
1. 去除不需要的行、列
2. 設定index為日期
3. 將逗點「,」刪除，讓table裡面的字串(str)變成數字(float or int)

```python
datestr + ' ' + df.index
```

```python
# 1. 去除不需要的行、列

import pandas as pd
import io
df = pd.read_csv(io.StringIO(res.text.replace("=", "")), header=1, index_col='時間')

df = df.dropna(how='all', axis=0).dropna(how='all', axis=1) #dropna = 刪除資料為空的行跟列、axis = 0(行)1(列)
df[:2]
```

```python
# 2. 設定index為日期
df.index = pd.to_datetime(datestr + ' ' + df.index)
#處理成電腦看得懂的格式之後

df['2019-01-02 9:00':'2019-1-2 9:00:20'] #用時間呼叫陣列
```

```python
# 3. 將逗點「,」刪除，讓table裡面的字串(str)變成數字(float or int)
df.dtypes #用dtypes查看資料型態 (object = 字串, float = 數字)
```

```python
df.astype(float) #astype 把資料全部轉成數字(但Error會顯示逗號無法轉)
```

```python
# 3. 將逗點「,」刪除，讓table裡面的字串(str)變成數字(float or int)
df = df.applymap(lambda s: float(str(s).replace(",", "")))
#applymap 把裡面字串中的物件str(s).replace(",", "")逗號取代掉
#在將取代後的float
#Lambda?
df[:10]
```

```python
#1~3 也可以這樣寫

def abc(e):
    ret = str(e)
    ret = ret.replace(",", "")
    ret = float(ret)
    return ret#float(str(e).replace(",", ""))

df = df.applymap(abc)
df.head()
```

### 全部加在一起

```python
# 時間物件
import datetime

# 下載網頁用的
import requests

# 資料處理
import pandas as pd

# 檔案串流
from io import StringIO


# 下載某天的資料

date = datetime.date(2018,1,3)
date_str = date.strftime('%Y%m%d')
res = requests.get("http://www.twse.com.tw/exchangeReport/MI_5MINS_INDEX?response=csv&date=" + 
                   date_str + "&_=1544020420045")

# 利用 pandas 將資料整理成表格

df = pd.read_csv(StringIO(res.text.replace("=","")), header=1, index_col='時間')

# 資料處理

df = df.dropna(how='all', axis=0).dropna(how='all', axis=1)
df.index = pd.to_datetime(date.strftime('%Y %m %d ') + pd.Series(df.index))
df = df.apply(lambda s: s.astype(str).str.replace(",", "").astype(float))

df.head()
```
